```python
car_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Computer Vision Masterclass/Cascades/cars.xml')
image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/car.jpg')
cv2_imshow(image)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
detections = car_detector.detectMultiScale(image_gray)
```

```python
detections = car_detector.detectMultiScale(image_gray,scaleFactor =1.05)
for (x,y,w,h) in detections:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
```

# FULL BODY DETECTOR

```python
full_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Computer Vision Masterclass/Cascades/fullbody.xml')
full_body =cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/people3.jpg')
#full_body = cv2.resize(full_body,(500,500))
full_body_gray = cv2.cvtColor(full_body,cv2.COLOR_BGR2GRAY)
cv2_imshow(full_body)
cv2_imshow(full_body_gray)
```

```python
full_body_detections = full_detector.detectMultiScale(full_body_gray, scaleFactor =2,minNeighbors =5,minSize = (50,50))
#full_body_detections
for (x,y,w,h) in full_body_detections:
   cv2.rectangle(full_body,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(full_body)
```
